```python
sentence = ["a = 3","b = 6","if a + b > 9 :","  b = a - 3","  a = b + a"]
def create_img(sentence,name):
    img = np.ones((1080,1920,3),np.uint8)*255
    img_ = img.copy()
#     for i,j  in enumerate(sentence):
#         cv2.putText(img_, j, (200, 100+150*i),font , 3, (0, 0, 0), lineType=cv2.LINE_AA)
    img_[:,1100:]=[200,200,200]
    plt.figure(figsize=(15,5))
#     plt.imshow(img_)
#     plt.show()
    img_ = Image.fromarray(img_)
    draw = ImageDraw.Draw(img_)
    for i,j  in enumerate(sentence):
        draw.text((200, 50+150*i), j, font = font, fill = (0,0,0))
    plt.imshow(img_)
    plt.show()
    img_.save(name)
#     img_.save("/Users/hasegawatakashikana/Downloads/img_.jpg")
```
